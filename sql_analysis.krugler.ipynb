I thought it would be interesting to find some empirical data about the one and only Yosemite that could prove useful for Texino. There is a huge percentage of campers in Yosemite that either use a van or RV. Without a doubt this is one of the most traveled spots in the world and if Texino could set up some sort of tour program with various stops to maximaze the consumers experience it would give them a stronger image in the eyes of the competition.

SQL used:
        CTE
        JOIN
        Group By

```python
%%sql 
WITH RV_ParkName_cte AS(
    SELECT ParkName, ParkRank AS RVcampingRank, Value, PercentOfTotal  
    FROM RV_Campers_2016
    WHERE ParkName LIKE 'Yosemite%'
),
Tent_ParkName_cte AS(
    SELECT ParkName, ParkRank AS TentCampingRank, Value, PercentOfTotal
    FROM Tent_Campers_2016
    WHERE ParkName LIKE 'Yosemite%'
),
Backcountry_ParkName_cte AS(
    SELECT ParkName, ParkRank AS BackcountryCampingRank, Value, PercentOfTotal
    FROM Backcountry_Campers_2016
    WHERE ParkName LIKE 'Yosemite%'
)
SELECT *
FROM RV_ParkName_cte R
JOIN Tent_ParkName_cte T
    ON R.ParkName = T.ParkName
JOIN Backcountry_ParkName_cte B
    ON T.ParkName = B.ParkName
LIMIT 1
    ;
```

```python
%%sql
SELECT *
FROM RVcampingYosemite   
 WHERE YearDate 
    BETWEEN 2013 AND 2016
ORDER BY YearDate DESC
```

```python
%%sql
SELECT *
FROM Tent_Campers_2016 AS Tent 
JOIN RV_Campers_2016 AS RV
    ON Tent.ParkName = RV.ParkName 
JOIN Backcountry_Campers_2016 AS Back
    ON RV.ParkName = Back.ParkName
GROUP BY Tent.ParkName 
LIMIT 10;
```

Customers often dont realize what they are getting themselves into with some tours. I wanted to create an easy to read guide for Texino to use as a base for their customers, giving them an accurate idea of the endeavour they are about to embark on.

```python
%%sql
SELECT TourName, TourType, TourDescription, TourDuration, 
AVG(TourDuration) OVER (PARTITION BY TourType) AS avg_duration_by_type,
    CASE
        WHEN TourDuration BETWEEN 0 AND 10 THEN 'Novice'
        WHEN TourDuration BETWEEN 11 AND 100 THEN 'Light Sweat'
        WHEN TourDuration BETWEEN 101 AND 599 THEN '2much'
        WHEN TourDuration BETWEEN 600 AND 1200 THEN 'You Asked 4 It'
    END AS DurationNotation
FROM Tours
GROUP BY TourDuration
ORDER BY TourDuration DESC
LIMIT 50;
```

I am aiming at creating a list for Texino to use to create an additional network to use during a time when finding customers is difficult. I found a lot of recreational areas but wanted to limit it to the ones that would be best suited for the adventures in a camper van. 

SQL used:
        View
        Window Function
        Subquery 
        Group By

```python
%%sql
CREATE OR REPLACE VIEW rec_area_parent_org AS
    SELECT  
            ParentOrgID,
            RecAreaName,
            RecAreaEmail,
            RecAreaPhone,
            Reservable,
            StayLimit,
            AVG(StayLimit) OVER() AS avg_stayLimit
          
    FROM RecAreas
    WHERE Keywords LIKE 'camp%'
    ORDER BY RecAreaName;

```

```python
%%sql
SELECT *
FROM rec_area_parent_org
WHERE ParentOrgID IN 
    (
        SELECT ParentOrgID
        FROM RecAreas
        WHERE Keywords LIKE 'camp%'
    )
GROUP BY RecAreaName
ORDER BY RecAreaName, ParentOrgID
LIMIT 25
;
```

```python
%%sql
WITH Rec_Area_cte AS(
    SELECT RecAreaName, RecAreaPhone, RecAreaEmail, StayLimit, Keywords, RecAreaMapURL
    FROM RecAreas
    WHERE Keywords LIKE 'camp%'
)
SELECT *
FROM Rec_Area_cte
LIMIT 10;


```

I wanted to find the peak dates when parks and recreational places had programs. I wanted to know the average attendence of a particular location. This could help inform and start run a marketing campaign for specific target areas that would be interested in Texino's services.
SQL used: 
            CTE
            Window Function
            Group By

```python
%%sql
WITH category_engagement_cte AS(
SELECT
    ProgramName,
    LocationName,
    Category,
    Attendence,
    Date
FROM engagement_tracker
GROUP BY ProgramName
)
SELECT 
    ProgramName,
    LocationName,
    Category,
    Attendence,
    Date,
    AVG(Attendence) OVER(PARTITION BY LocationName) AS avg_attendence
FROM category_engagement_cte
GROUP BY Category
ORDER BY Date, Attendence DESC
LIMIT 15;
```
